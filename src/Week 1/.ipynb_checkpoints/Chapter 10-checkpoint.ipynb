## 10.1 Matplotlib

### 10.1.1 Matplotlib

networkx.drawing.nx_pylab.draw<br><br>
draw(G, pos=None, ax=None, **kwds) Draw the graph G with Matplotlib.<br>
Draw the graph as a simple representation with no node labels or edge labels and using the full Matplotlib figure area and no axis labels by default. See draw_networkx() for more full-featured drawing that allows title, axis labels etc.<br><br>
Parameters<br>
• G (graph) – A networkx graph<br>
• pos (dictionary, optional) – A dictionary with nodes as keys and positions as values. If not specified a spring layout positioning will be computed. See networkx.drawing. layout for functions that compute node positions.<br>
• ax (Matplotlib Axes object, optional) – Draw the graph in specified Matplotlib axes.<br>
• kwds (optional keywords) – See networkx.draw_networkx() for a description of optional
keywords.

```python

import networkx as nx

G = nx.dodecahedral_graph()
nx.draw(G=G)
nx.draw(G=G, pos=nx.spring_layout(G=G), with_labels=True)
```


networkx.drawing.nx_pylab.draw_networkx<br><br>
draw_networkx(G, pos=None, arrows=True, with_labels=True, **kwds)<br>
Draw the graph G using Matplotlib.<br>
Draw the graph with Matplotlib with options for node positions, labeling, titles, and many other drawing fea- tures. See draw() for simple drawing without labels or axes.<br><br>
Parameters<br>
• G (graph) – A networkx graph<br>
• pos (dictionary, optional) – A dictionary with nodes as keys and positions as values. If not specified a spring layout positioning will be computed. See networkx.drawing. layout for functions that compute node positions.<br>
• arrows (bool, optional (default=True)) – For directed graphs, if True draw arrowheads.<br><br>
Note: Arrows will be the same color as edges.<br>
• arrowstyle (str, optional (default=’-|>’)) – For directed graphs, choose the style of the arrowsheads. See :py:class: matplotlib.patches.ArrowStyle for more options.<br>
• arrowsize (int, optional (default=10)) – For directed graphs, choose the size of the arrow head head’s length and width. See :py:class: matplotlib.patches. FancyArrowPatch for attribute mutation_scale for more info.<br>
• with_labels (bool, optional (default=True)) – Set to True to draw labels on the nodes.<br>
• ax (Matplotlib Axes object, optional) – Draw the graph in the specified Matplotlib axes.<br>
• nodelist (list, optional (default G.nodes())) – Draw only specified nodes<br>
• edgelist (list, optional (default=G.edges())) – Draw only specified edges<br>
• node_size (scalar or array, optional (default=300)) – Size of nodes. If an array is specified it must be the same length as nodelist.<br>
• node_color (color or array of colors (default=’#1f78b4’)) – Node color. Can be a single color or a sequence of colors with the same length as nodelist. Color can be string, or rgb (or rgba) tuple of floats from 0-1. If numeric values are specified they will be mapped to colors using the cmap and vmin,vmax parameters. See matplotlib.scatter for more details.<br>
• node_shape (string, optional (default=’o’)) – The shape of the node. Specification is as matplotlib.scatter marker, one of ‘so^>v<dph8’.<br>
• alpha (float, optional (default=None)) – The node and edge transparency<br>
• cmap (Matplotlib colormap, optional (default=None)) – Colormap for mapping intensities
of nodes<br>
• vmin,vmax (float, optional (default=None)) – Minimum and maximum for node colormap scaling<br>
• linewidths ([None | scalar | sequence]) – Line width of symbol border (default =1.0)<br>
• width (float, optional (default=1.0)) – Line width of edges<br>
• edge_color (color or array of colors (default=’k’)) – Edge color. Can be a single color or a sequence of colors with the same length as edgelist. Color can be string, or rgb (or rgba) tuple of floats from 0-1. If numeric values are specified they will be mapped to colors using the edge_cmap and edge_vmin,edge_vmax parameters.<br>
• edge_cmap (Matplotlib colormap, optional (default=None)) – Colormap for mapping in- tensities of edges<br>
• edge_vmin,edge_vmax (floats, optional (default=None)) – Minimum and maximum for edge colormap scaling<br>
• style (string, optional (default=’solid’)) – Edge line style (solid|dashed|dotted,dashdot)<br>
• labels (dictionary, optional (default=None)) – Node labels in a dictionary keyed by node of text labels<br>
• font_size (int, optional (default=12)) – Font size for text labels<br>
• font_color (string, optional (default=’k’ black)) – Font color string<br>
• font_weight (string, optional (default=’normal’)) – Font weight<br>
• font_family (string, optional (default=’sans-serif’)) – Font family<br>
• label (string, optional) – Label for graph legend<br><br>

Notes<br>
For directed graphs, arrows are drawn at the head end. Arrows can be turned off with keyword arrows=False.


```python

import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

G = nx.dodecahedral_graph()
nx.draw(G=G)
nx.draw(G=G, pos=nx.spring_layout(G=G), with_labels=True)
limits = plt.axis('off')
```

```python

import networkx as nx
import numpy as np

G = nx.hexagonal_lattice_graph(m=4, n=2)
nodes_remove = [n[0] == 1 and n[1] in range(2, 8) for n in G]
print([(n, remove) for n, remove in zip(G.nodes, nodes_remove)])
nodelist = [n for n, remove in zip(G.nodes, nodes_remove) if not remove]
print('-' * 125)
print(nodelist)

nx.draw_networkx(G=G, pos=nx.spring_layout(G=G), with_labels=True, nodelist=nodelist)
```

```python

import networkx as nx

G = nx.path_graph(n=10, create_using=nx.MultiDiGraph)
node_size = [(n + 1) * 100 for n in G]
print(node_size)

nx.draw_networkx(G=G,
              with_labels=True,
              node_shape='s',
              node_size=node_size,
              alpha=0.5,
              font_family='Apple Chancery',
              font_size=24,
              font_weight=5,
              font_color='red')
```

networkx.drawing.nx_pylab.draw_networkx_nodes<br><br>
draw_networkx_nodes(G, pos, nodelist=None, node_size=300, node_color='#1f78b4', node_shape='o', alpha=None, cmap=None, vmin=None, vmax=None, ax=None, linewidths=None, edgecolors=None, label=None, **kwds)<br>
Draw the nodes of the graph G.<br>
This draws only the nodes of the graph G.<br><br>
Parameters<br>
• G (graph) – A networkx graph<br>
• pos (dictionary) – A dictionary with nodes as keys and positions as values. Positions should be sequences of length 2.<br>
• ax (Matplotlib Axes object, optional) – Draw the graph in the specified Matplotlib axes.<br>
• nodelist (list, optional) – Draw only specified nodes (default G.nodes())<br>
• node_size (scalar or array) – Size of nodes (default=300). If an array is specified it must be the same length as nodelist.<br>
• node_color (color or array of colors (default=’#1f78b4’)) – Node color. Can be a single color or a sequence of colors with the same length as nodelist. Color can be string, or rgb (or rgba) tuple of floats from 0-1. If numeric values are specified they will be mapped to colors using the cmap and vmin,vmax parameters. See matplotlib.scatter for more details.<br>
• node_shape (string) – The shape of the node. Specification is as matplotlib.scatter marker, one of ‘so^>v<dph8’ (default=’o’).<br>
• alpha (float or array of floats) – The node transparency. This can be a single alpha value (default=None), in which case it will be applied to all the nodes of color. Otherwise, if it is an array, the elements of alpha will be applied to the colors in order (cycling through alpha multiple times if necessary).<br>
• cmap (Matplotlib colormap) – Colormap for mapping intensities of nodes (default=None)<br>
• vmin,vmax (floats) – Minimum and maximum for node colormap scaling (default=None)<br>
• linewidths ([None | scalar | sequence]) – Line width of symbol border (default =1.0)<br>
• edgecolors ([None | scalar | sequence]) – Colors of node borders (default = node_color)<br>
• label ([None| string]) – Label for legend<br>
• min_source_margin (int, optional (default=0)) – The minimum margin (gap) at the begin-
ing of the edge at the source.<br>
• min_target_margin (int, optional (default=0)) – The minimum margin (gap) at the end of the edge at the target.<br>
Returns PathCollection of the nodes. Return type matplotlib.collections.PathCollection

```python

import networkx as nx

G = nx.dodecahedral_graph()
nodes = nx.draw_networkx_nodes(G=G, pos=nx.spring_layout(G=G))
```

```python

import networkx as nx

G = nx.DiGraph()
G.add_edges_from(ebunch_to_add=[(1, 2), (1, 2), (2, 3)])
arcs = nx.draw_networkx_edges(G=G, pos=nx.spring_layout(G=G))
alphas = [0.3, 0.4, 0.5]
for i, arc in enumerate(arcs):
    arc.set_alpha(alphas[i])
```

Returns dict of labels keyed on the edges Return type dict

```python

import networkx as nx
import numpy as np

G = nx.dodecahedral_graph()
for e in G.edges:
    """Append random edge labels to each edge"""
    G.edges[e]['weight'] = np.round(np.random.rand(), 2)
    G.edges[e]['char'] = np.random.randint(32, 126)

edge_labels = nx.draw_networkx_edge_labels(G=G, pos=nx.spring_layout(G=G))
```

networkx.drawing.nx_pylab.draw_spectral<br><br>

Draw the graph G with a spectral 2D layout.<br>

Parameters<br><br>

• kwargs (optional keywords) – See networkx.draw_networkx() for a description of optional keywords, with the exception of the pos parameter which is not used by this function.

```python

import networkx as nx

G = nx.complete_graph(n=10)
s = nx.draw_spectral(G=G)
```
