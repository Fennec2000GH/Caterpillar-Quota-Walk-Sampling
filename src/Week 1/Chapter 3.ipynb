## Graph Traversal

dfs_edges(G, source=None, depth_limit=None)<br>
Iterate over edges in a depth-first-search (DFS).<br>
Perform a depth-first-search over the nodes of G and yield the edges in order. This may not generate all edges in G (see edge_dfs).<br><br>
Parameters<br>
• G (NetworkX graph)<br>
• source(node,optional)–Specifystartingnodefordepth-firstsearchandreturnedgesinthe component reachable from source.<br>
• depth_limit (int, optional (default=len(G))) – Specify the maximum search depth. Returns edges – A generator of edges in the depth-first-search.<br>
Return type generator

```python

import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.traversal import dfs_edges
%matplotlib inline

G = nx.path_graph(n=5)
nx.draw(G=G)
plt.title(label='Path Graph')
plt.show()

print(list(dfs_edges(G=G))) # depth-first traversal of edges
print(list(dfs_edges(G=G, source=0, depth_limit=2))) # DFS of edges limited to search depth of 2
```

networkx.algorithms.traversal.depth_first_search.dfs_predecessors<br><br>
dfs_predecessors(G, source=None, depth_limit=None)<br>
Returns dictionary of predecessors in depth-first-search from source.<br><br>
Parameters<br>
• G (NetworkX graph)<br>
• source (node, optional) – Specify starting node for depth-first search.<br>
• depth_limit (int, optional (default=len(G))) – Specify the maximum search depth. Returns pred – A dictionary with nodes as keys and predecessor nodes as values.<br>
Return type dict

```python

import matplotlib.pyplot as plt
import networkx as nx
from networkx.algorithms.traversal import dfs_predecessors
%matplotlib inline

G = nx.path_graph(n=5, create_using=nx.DiGraph)
nx.draw(G=G)
plt.title(label='Directed Path Graph')
plt.show()

print(dfs_predecessors(G=G, source=0))
print(dfs_predecessors(G=G, source=0, depth_limit=2))
```
