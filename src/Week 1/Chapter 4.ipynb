networkx.classes.function.degree<br><br>
degree(G, nbunch=None, weight=None)<br>
Returns a degree view of single node or of nbunch of nodes. If nbunch is omitted, then return degrees of all nodes.

```python

import matplotlib.pyplot as plt
import networkx as nx
from networkx.generators import complete_graph
%matplotlib inline

G = complete_graph(n=10)
degree_list = nx.degree(G=G)
print(degree_list)
print([item[1] for item in degree_list])
assert [item[1] for item in degree_list] == [9] * len(degree_list)
nx.draw(G=G)
plt.show()
```

networkx.classes.function.degree_histogram<br><br>
degree_histogram(G)<br>
Returns a list of the frequency of each degree value.<br>
Parameters G (Networkx graph)  A graph<br>
Returns hist  A list of frequencies of degrees. The degree values are the index in the list.<br>
Return type list<br><br>
Notes<br>
Note: the bins are width one, hence len(list) can be large (Order(number_of_edges))

```python

import matplotlib.pyplot as plt
import networkx as nx
from networkx.generators.random_graphs import watts_strogatz_graph
%matplotlib inline

G = watts_strogatz_graph(n=10, k=5, p=0.5)
degree_list = nx.degree_histogram(G=G)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
nx.draw(G=G, ax=ax1, with_labels=True)
ax2.hist(x=degree_list, color='green', facecolor='blue', edgecolor='white')
ax2.set_title(label='Degree Histogram', size=24)
ax2.set_xlabel(xlabel='Degree')
ax2.set_ylabel(ylabel='Number of Nodes')
plt.show()
```

networkx.classes.function.is_empty<br><br>
is_empty(G)<br>
Returns True if G has no edges.<br><br>
Parameters G (graph)  A NetworkX graph. Returns True if G has no edges, and False otherwise.<br>
Return type bool<br>

Notes<br>
An empty graph can have nodes but not edges. The empty graph with zero nodes is known as the null graph. This is an () operation where n is the number of nodes in the graph.


```python

import matplotlib.pyplot as plt
import networkx as nx
from networkx.generators.geometric import soft_random_geometric_graph
%matplotlib inline

G = soft_random_geometric_graph(n=10, radius=10)
empty_1 = nx.empty_graph(n=10)
empty_2 = nx.empty_graph()

print(nx.is_empty(G=G))
print(nx.is_empty(G=empty_1))
print(nx.is_empty(G=empty_2))

fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, figsize=(30, 10))
fig.suptitle(t='is_empty Example', size=36)
nx.draw(G=G, ax=ax1, with_labels=True)
ax1.set_title(label='Non-Empty Graph', size=24)
nx.draw(G=empty_1, ax=ax2, with_label=True)
ax2.set_title(label='Empty Graph', size=24)
nx.draw(G=empty_2, ax=ax3, with_labels=True)
ax3.set_title(label='Empty Graph', size=24)
plt.show()
```

networkx.classes.function.subgraph_view<br><br>
subgraph_view(G, filter_node=<function no_filter>, filter_edge=<function no_filter>)<br>
View of G applying a filter on nodes and edges.<br>
subgraph_view provides a read-only view of the input graph that excludes nodes and edges based on the outcome of two filter functions filter_node and filter_edge.<br>
The filter_node function takes one argument  the node  and returns True if the node should be included in the subgraph, and False if it should not be included.<br>
The filter_edge function takes two arguments  the nodes describing an edge  and returns True if the edge should be included in the subgraph, and False if it should not be included.<br>
Both node and edge filter functions are called on graph elements as they are queried, meaning there is no up-front cost to creating the view.<br><br>
Parameters<br>
 G (networkx.Graph)  A directed/undirected graph/multigraph<br>
 filter_node (callable, optional)  A function taking a node as input, which returns True if the node should appear in the view.<br>
 filter_edge (callable, optional)  A function taking as input the two nodes describing an edge, which returns True if the edge should appear in the view.<br>
Returns graph  A read-only graph view of the input graph. Return type networkx.Graph

```python

import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

G1 = nx.path_graph(n=10)
def edge_filter(a, b):
    return False
G2 = nx.subgraph_view(G=G1, filter_node=lambda n: n >= 4 and n <= 8, filter_edge=edge_filter)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))
fig.suptitle(t='subgraph Example', size=36)
nx.draw(G=G1, ax=ax1, with_labels=True)
ax1.set_title(label='Original', size=24)
nx.draw(G=G2, ax=ax2, with_labels=True)
ax2.set_title(label='Subgraph', size=24)
plt.show()

```

networkx.classes.function.induced_subgraph<br><br>
induced_subgraph(G, nbunch)<br>
Returns a SubGraph view of G showing only nodes in nbunch.<br>
The induced subgraph of a graph on a set of nodes N is the graph with nodes N and edges from G which have both ends in N.<br><br>
Parameters<br>
 G (NetworkX Graph)<br>
 nbunch (node, container of nodes or None (for all nodes))<br>
Returns subgraph  A read-only view of the subgraph in G induced by the nodes. Changes to the graph G will be reflected in the view.<br>
Return type SubGraph View<br><br>
Notes<br>
To create a mutable subgraph with its own copies of nodes edges and attributes use subgraph.copy() or Graph(subgraph). For an inplace reduction of a graph to a subgraph you can remove nodes: G.remove_nodes_from(n in G if n not in set(nbunch)). If you are going to compute subgraphs of your subgraphs you could end up with a chain of views that can be very slow once the chain has about 15 views in it. If they are all induced subgraphs, you can short-cut the chain by making them all subgraphs of the original graph. The graph class method G.subgraph does this when G is a subgraph. In contrast, this function allows you to choose to build chains or not, as you wish. The returned subgraph is a view on G.

```python

import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

G1 = nx.path_graph(n=10)
G2 = nx.induced_subgraph(G=G1, nbunch=0)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(30, 10))
fig.suptitle(t='induced_subgraph Example', size=36)
nx.draw(G=G1, ax=ax1, with_labels=True)
ax1.set_title(label='Original', size=24)
nx.draw(G=G2, ax=ax2, with_labels=True)
ax2.set_title(label='Induced Subgraph', size=24)
plt.show()

```

networkx.classes.function.restricted_view<br><br>
restricted_view(G, nodes, edges)<br>
Returns a view of G with hidden nodes and edges.<br>
The resulting subgraph filters out node nodes and edges edges. Filtered out nodes also filter out any of their edges.<br><br>
Parameters<br>
 G (NetworkX Graph)<br>
 nodes (iterable)  An iterable of nodes. Nodes not present in G are ignored.<br>
 edges (iterable)  An iterable of edges. Edges not present in G are ignored.<br>
Returns subgraph  A read-only restricted view of G filtering out nodes and edges. Changes to G are reflected in the view.<br>
Return type SubGraph View

```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

G1 = nx.path_graph(n=10)
print(list(G1.edges)[::2])
G2 = nx.restricted_view(G=G1, nodes=[], edges=list(G1.edges)[::2])

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
fig.suptitle(t='restricted_view Example', size=36)
nx.draw(G=G1, ax=ax1, with_labels=True)
ax1.set_title(label='Original', size=24)
nx.draw(G=G2, ax=ax2, with_labels=True)
ax2.set_title(label='Restricted', size=24)
plt.show()
```

networkx.classes.function.edge_subgraph<br><br>
edge_subgraph(G, edges)<br>
Returns a view of the subgraph induced by the specified edges.<br>
The induced subgraph contains each edge in edges and each node incident to any of those edges.<br><br>
Parameters<br>
 G (NetworkX Graph)<br>
 edges (iterable)  An iterable of edges. Edges not present in G are ignored.<br>
Returns subgraph  A read-only edge-induced subgraph of G. Changes to G are reflected in the
view.<br>
Return type SubGraph View<br><br>
Notes<br>
To create a mutable subgraph with its own copies of nodes edges and attributes use subgraph.copy() or Graph(subgraph). If you create a subgraph of a subgraph recursively you can end up with a chain of subgraphs that becomes very slow with about 15 nested subgraph views. Luckily the edge_subgraph filter nests nicely so you can use the original graph as G in this function to avoid chains. We do not rule out chains programmatically so that odd cases like an edge_subgraph of a restricted_view can be created.

```python

import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline

G1 = nx.path_graph(n=10)
print(list(G1.edges)[::2])
G2 = nx.edge_subgraph(G=G1, edges=list(G1.edges)[::2])

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 10))
fig.suptitle(t='edge_subgraph Example', size=36)
nx.draw(G=G1, ax=ax1, with_labels=True)
ax1.set_title(label='Original', size=24)
nx.draw(G=G2, ax=ax2, with_labels=True)
ax2.set_title(label='Edge-Induced Subgraph', size=24)
plt.show()
```
